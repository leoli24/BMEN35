```python
from numpy.core.fromnumeric import mean
import numpy as np
import matplotlib as plt

x = np.arange(0, 5, 5/1000)
y = list(map((lambda x:pow(x,2)), x))


def mean_hundred(y):
 numbers = np.random.randint(0, 1000, 100)
 m = []
 for n in numbers:
    m.append((y[n]))
 
 return np.mean(m)
 
y_mean = mean_hundred(y)

print(y_mean)
```

```python
clt = []
for i in range(50000):
 clt.append(mean_hundred(y))

print(clt)

plt.pyplot.hist(clt)
```

The plot is showing a normal distribution.
